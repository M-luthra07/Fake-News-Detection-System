```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import tensorflow as tf
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, classification_report,
                           confusion_matrix, roc_curve, auc,
                           precision_score, recall_score, f1_score)
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.utils import class_weight
from transformers import BertTokenizer, TFBertModel
from keras.layers import Input, LSTM, Dense, Layer, Bidirectional, Dropout
from keras.models import Model
from keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings('ignore')
```

```python
# Hyperparameters
MAX_LENGTH = 128
BATCH_SIZE = 32
EPOCHS = 10
DROPOUT_RATE = 0.3
LSTM_UNITS = 128
# Text cleaning function
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\@\w+|\#', '', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.strip()
```

```python
# Loading Dataset
datasets = {
    "politifact_fake": "https://raw.githubusercontent.com/KaiDMML/FakeNewsNet/master/dataset/politifact_fake.csv",
    "politifact_real": "https://raw.githubusercontent.com/KaiDMML/FakeNewsNet/master/dataset/politifact_real.csv",
    "gossipcop_fake": "https://raw.githubusercontent.com/KaiDMML/FakeNewsNet/master/dataset/gossipcop_fake.csv",
    "gossipcop_real": "https://raw.githubusercontent.com/KaiDMML/FakeNewsNet/master/dataset/gossipcop_real.csv"
}

dfs = []
for name, url in datasets.items():
    try:
        df = pd.read_csv(url)
        df["label"] = 1 if "fake" in name else 0
        dfs.append(df)
    except Exception as e:
        print(f"Error loading {name}: {e}")

data = pd.concat(dfs, ignore_index=True)
print(f"✅ Dataset loaded successfully! {len(data)} records")

# Clean and augment short texts
data['cleaned_title'] = data['title'].apply(clean_text)
data['text_length'] = data['cleaned_title'].apply(len)

# Identify and duplicate short texts to balance the dataset
short_texts = data[data['text_length'] < 10]
data = pd.concat([data, short_texts], ignore_index=True)

texts = data['cleaned_title'].values
labels = data['label'].values
```

```python
# BERT Tokenization
print("Tokenizing text...")
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
inputs = tokenizer(texts.tolist(), return_tensors='tf', padding='max_length', truncation=True, max_length=MAX_LENGTH)

# BERT Model
print("Loading BERT model...")
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
bert_model.trainable = False

class BertLayer(Layer):
    def __init__(self, bert_model, **kwargs):
        super(BertLayer, self).__init__(**kwargs)
        self.bert_model = bert_model

    def call(self, inputs):
        input_ids, attention_mask = inputs
        bert_output = self.bert_model(input_ids, attention_mask=attention_mask)[0]
        return bert_output
```

```python
# Improved Model Architecture
def create_improved_model():
    input_ids = Input(shape=(MAX_LENGTH,), dtype=tf.int32, name="input_ids")
    attention_mask = Input(shape=(MAX_LENGTH,), dtype=tf.int32, name="attention_mask")

    bert_output = BertLayer(bert_model)([input_ids, attention_mask])

    # Enhanced architecture
    x = Bidirectional(LSTM(LSTM_UNITS, return_sequences=True))(bert_output)
    x = Dropout(DROPOUT_RATE)(x)
    x = Bidirectional(LSTM(LSTM_UNITS//2))(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(DROPOUT_RATE)(x)
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=[input_ids, attention_mask], outputs=output)
    model.compile(
        optimizer='Adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

print("Creating improved LSTM model...")
model = create_improved_model()
model.optimizer.lr = 2e-5

# Class weights for imbalance
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(labels),
    y=labels
)
class_weights = dict(enumerate(class_weights))
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
print("Training model...")
history = model.fit(
    x={'input_ids': inputs['input_ids'], 'attention_mask': inputs['attention_mask']},
    y=labels,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    validation_split=0.2,
    callbacks=[early_stopping],
    class_weight=class_weights
)
```

```python
# Feature extraction
print("Extracting features...")
feature_extractor = Model(inputs=model.inputs, outputs=model.layers[-3].output)
features = feature_extractor.predict({'input_ids': inputs['input_ids'], 'attention_mask': inputs['attention_mask']})

# SVM Model
print("Training SVM...")
svm_model = make_pipeline(
    StandardScaler(),
    SVC(kernel='rbf', C=1.0, gamma='scale', probability=True)
)
svm_model.fit(features, labels)
```

```python
# Evaluation Functions
def evaluate_lstm():
    predictions = (model.predict({'input_ids': inputs['input_ids'],
                                'attention_mask': inputs['attention_mask']}) > 0.5).astype(int)
    return {
        'Accuracy': accuracy_score(labels, predictions),
        'Precision': precision_score(labels, predictions),
        'Recall': recall_score(labels, predictions),
        'F1-Score': f1_score(labels, predictions)
    }
def evaluate_svm():
    predictions = svm_model.predict(features)
    return {
        'Accuracy': accuracy_score(labels, predictions),
        'Precision': precision_score(labels, predictions),
        'Recall': recall_score(labels, predictions),
        'F1-Score': f1_score(labels, predictions)
    }
def evaluate_hybrid():
    # Get all predictions and probabilities at once
    all_proba = svm_model.predict_proba(features)
    all_preds = svm_model.predict(features)

    # Precompute text features
    cleaned_texts = [clean_text(text).lower() for text in texts]
    text_lengths = np.array([len(t.split()) for t in cleaned_texts])
    has_names = np.array([any(name in t for name in ['sid', 'miller', 'trump', 'biden'])
                         for t in cleaned_texts])

    # Create conditions
    short_text = text_lengths < 3
    low_confidence = np.max(all_proba, axis=1) < 0.7
    special_case = short_text & low_confidence

    # Apply hybrid rules
    predictions = np.where(
        special_case & has_names, 1,  # Mark as fake
        np.where(
            special_case & ~has_names, 0,  # Mark as real
            all_preds  # Default to SVM prediction
        )
    )

    return {
        'Accuracy': accuracy_score(labels, predictions),
        'Precision': precision_score(labels, predictions),
        'Recall': recall_score(labels, predictions),
        'F1-Score': f1_score(labels, predictions)
    }
```

```python
4358# Generate Metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score # Import precision_score, recall_score, and f1_score

print("\nEvaluating models...")
lstm_metrics = evaluate_lstm()
svm_metrics = evaluate_svm()
hybrid_metrics = evaluate_hybrid()

# Create Results DataFrame
results = pd.DataFrame({
    'Model': ['LSTM', 'SVM', 'Hybrid'],
    'Accuracy': [lstm_metrics['Accuracy'], svm_metrics['Accuracy'], hybrid_metrics['Accuracy']],
    'Precision': [lstm_metrics['Precision'], svm_metrics['Precision'], hybrid_metrics['Precision']],
    'Recall': [lstm_metrics['Recall'], svm_metrics['Recall'], hybrid_metrics['Recall']],
    'F1-Score': [lstm_metrics['F1-Score'], svm_metrics['F1-Score'], hybrid_metrics['F1-Score']]
})

# Generate Classification Reports
print("\nLSTM Model Classification Report:")
print(classification_report(labels, (model.predict({'input_ids': inputs['input_ids'],
                                                 'attention_mask': inputs['attention_mask']}) > 0.5).astype(int),
                           target_names=['Real', 'Fake']))

print("\nSVM Model Classification Report:")
print(classification_report(labels, svm_model.predict(features),
                           target_names=['Real', 'Fake']))
```

```python
def plot_model_comparison(results):
    """Generate bar chart comparing model performance metrics"""
    # Convert metrics to percentages
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score']
    results[metrics] = results[metrics] * 100

    plt.figure(figsize=(10, 6))
    ax = results.plot(x='Model', y=metrics, kind='bar', width=0.8,
                     color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

    plt.title('Model Performance Comparison', pad=20)
    plt.ylabel('Score (%)')
    plt.ylim(0, 105)
    plt.xticks(rotation=0)

    # Add value labels
    for p in ax.patches:
        ax.annotate(f"{p.get_height():.1f}%",
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center',
                   xytext=(0, 5),
                   textcoords='offset points')

    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.savefig('model_comparison.png', bbox_inches='tight')
    plt.show()
```

```python
plot_model_comparison(results)
```

```python
def plot_training_history(history):
    """Plot training and validation accuracy/loss curves"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # Accuracy plot
    ax1.plot(history.history['accuracy'], label='Training', color='#1f77b4')
    ax1.plot(history.history['val_accuracy'], label='Validation', color='#ff7f0e')
    ax1.set_title('Accuracy Curve')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid(linestyle='--', alpha=0.5)

    # Loss plot
    ax2.plot(history.history['loss'], label='Training', color='#1f77b4')
    ax2.plot(history.history['val_loss'], label='Validation', color='#ff7f0e')
    ax2.set_title('Loss Curve')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid(linestyle='--', alpha=0.5)

    plt.suptitle('Training History', y=1.02)
    plt.tight_layout()
    plt.savefig('training_history.png')
    plt.show()
```

```python
print("Generating visualizations...")
plot_training_history(history)
```

```python
def plot_class_distribution(labels):
    """Plot pie chart of class distribution"""
    counts = pd.Series(labels).value_counts()

    plt.figure(figsize=(6, 6))
    plt.pie(counts,
            labels=['Real', 'Fake'],
            colors=['#66b3ff', '#ff9999'],
            autopct='%1.1f%%',
            startangle=90,
            explode=(0.05, 0),
            shadow=True)
    plt.title('Class Distribution')
    plt.tight_layout()
    plt.savefig('class_distribution.png')
    plt.show()
```
